```python
import os
import json
```

```python
def build_tree(path):
    tree = {}
    for entry in os.listdir(path):
        full_path = os.path.join(path, entry)
        if os.path.isdir(full_path):
            tree[entry] = build_tree(full_path)
        else:
            tree[entry] = None  # 또는 파일 크기: os.path.getsize(full_path)
    return tree
```

```python

root_path = "C:\\Users\\pawne\\Desktop\\수업"  # 여기에 실제 경로 입력
tree = {os.path.basename(root_path): build_tree(root_path)}

# JSON 형식으로 출력
print(json.dumps(tree, indent=2, ensure_ascii=False))
```

```python
with open("디렉터리구조.json", "w", encoding="utf-8") as f:
    json.dump(tree, f, indent=2, ensure_ascii=False)
```
